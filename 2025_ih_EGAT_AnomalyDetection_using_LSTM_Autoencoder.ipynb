```python
# Lab: Anomaly Detection with LSTM Autoencoder
# T4 GPU Environment - Google Colab

# ติดตั้ง packages ที่จำเป็น (รันเฉพาะครั้งแรก)
!pip install -q tensorflow numpy pandas matplotlib seaborn scikit-learn

# นำเข้าไลบรารีที่จำเป็น
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Input, Dense, LSTM, RepeatVector, TimeDistributed
from tensorflow.keras.callbacks import EarlyStopping
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import precision_recall_fscore_support, confusion_matrix

# Confirm GPU is available for TensorFlow
print("TensorFlow version:", tf.__version__)
print("GPU Available:", tf.config.list_physical_devices('GPU'))

# Set seed for reproducibility
np.random.seed(42)
tf.random.set_seed(42)

# ============================================================================
# PART 1: Generate Time Series Data
# ============================================================================

def create_power_production_data(start_date='2020-01-01', end_date='2022-12-31'):
    """
    สร้างชุดข้อมูลจำลองการผลิตไฟฟ้าที่มีองค์ประกอบหลักของข้อมูล Time Series
    - Trend: การเติบโตของความต้องการไฟฟ้า 3% ต่อปี
    - Seasonality: รายปี, รายวัน, รายสัปดาห์
    - Cyclical: วัฏจักรทางเศรษฐกิจประมาณ 3 ปี
    - Irregular: ความผันผวนทั่วไป
    - Special Events: เหตุการณ์พิเศษ (ซ่อมบำรุง, เหตุฉุกเฉิน)
    """
    # สร้างช่วงวันที่
    date_range = pd.date_range(start=start_date, end=end_date, freq='H')
    df = pd.DataFrame(index=date_range)

    # 1. แนวโน้ม (Trend) - การเติบโตของความต้องการไฟฟ้า 3% ต่อปี
    days = (df.index - df.index[0]).total_seconds() / (24*60*60)
    daily_growth = (1.03)**(1/365)  # การเติบโต 3% ต่อปี
    trend = 1000 * (daily_growth ** (days))

    # 2. ฤดูกาล (Seasonality)
    # ฤดูกาลรายปี - ความต้องการไฟฟ้าเพิ่มขึ้นในฤดูร้อนและฤดูหนาว
    yearly_seasonality = 150 * np.sin(2 * np.pi * days / 365)

    # ฤดูกาลรายวัน - สูงในช่วงกลางวัน ต่ำในช่วงกลางคืน
    hour_of_day = df.index.hour
    daily_pattern = 100 * np.sin(np.pi * hour_of_day / 12)

    # ฤดูกาลรายสัปดาห์ - ต่ำในวันหยุดสุดสัปดาห์
    day_of_week = df.index.dayofweek
    weekly_pattern = np.where(day_of_week >= 5, -80, 0)  # ลด 80 หน่วยในวันเสาร์-อาทิตย์

    # 3. วัฏจักร (Cyclical) - วัฏจักรทางเศรษฐกิจประมาณ 3 ปี
    cycle_period = 3 * 365  # วัฏจักร 3 ปี
    cyclical = 70 * np.sin(2 * np.pi * days / cycle_period)

    # 4. Irregular Component - ความผันผวนทั่วไปและเหตุการณ์พิเศษ
    np.random.seed(42)  # ตั้งค่า seed สำหรับการสร้างค่าแบบสุ่ม
    irregular = np.random.normal(0, 30, len(df))

    # เพิ่มเหตุการณ์พิเศษบางวัน (เช่น ไฟดับหรือซ่อมบำรุงโรงไฟฟ้า)
    special_events = pd.Series(0, index=df.index)

    # สมมติว่ามีการซ่อมบำรุงโรงไฟฟ้าในบางช่วง
    maintenance_periods = [
        ('2021-04-10', '2021-04-15'),
        ('2022-09-05', '2022-09-10')
    ]

    for start, end in maintenance_periods:
        mask = (df.index >= start) & (df.index <= end)
        special_events[mask] = -200  # ลดการผลิตลง 200 หน่วย

    # สมมติว่ามีเหตุการณ์ฉุกเฉินทำให้โรงไฟฟ้าหยุดทำงานชั่วคราว
    emergency_dates = ['2021-07-15', '2022-11-30']
    for date in emergency_dates:
        # ส่งผลกระทบ 48 ชั่วโมง
        start_date = pd.Timestamp(date)
        end_date = start_date + pd.Timedelta(hours=48)
        mask = (df.index >= start_date) & (df.index < end_date)
        special_events[mask] = -350  # ลดการผลิตลง 350 หน่วย

    # รวมองค์ประกอบทั้งหมด
    df['power_production'] = trend + yearly_seasonality + daily_pattern + weekly_pattern + cyclical + irregular + special_events

    # ปรับให้ค่าต่ำสุดไม่น้อยกว่า 100 (กำลังการผลิตขั้นต่ำ)
    df['power_production'] = df['power_production'].clip(lower=100)

    # เพิ่มคอลัมน์ข้อมูลเวลาที่ง่ายต่อการวิเคราะห์
    df['hour'] = df.index.hour
    df['day'] = df.index.day
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['day_of_week'] = df.index.dayofweek
    df['weekend'] = df['day_of_week'].apply(lambda x: 1 if x >= 5 else 0)

    return df

# ============================================================================
# PART 2: Add Additional Anomalies for Testing
# ============================================================================

def add_anomalies(df, n_point_anomalies=10, n_collective_anomalies=3, n_contextual_anomalies=5):
    """
    เพิ่มความผิดปกติ 3 ประเภทเข้าไปในข้อมูล:
    1. Point Anomalies: ค่าที่สูงหรือต่ำผิดปกติในจุดเดียว
    2. Collective Anomalies: ค่าที่มีรูปแบบผิดปกติต่อเนื่องหลายจุด
    3. Contextual Anomalies: ค่าที่ผิดปกติเมื่อพิจารณาบริบท (เช่น ช่วงเวลา)
    """
    # สร้าง DataFrame ใหม่เพื่อไม่เปลี่ยนแปลงข้อมูลเดิม
    df_anomaly = df.copy()

    # สร้างคอลัมน์บ่งชี้ความผิดปกติ
    df_anomaly['anomaly'] = 0  # 0 = ปกติ, 1 = ผิดปกติ
    df_anomaly['anomaly_type'] = 'normal'  # ประเภทของความผิดปกติ

    # ตำแหน่งที่มีเหตุการณ์พิเศษอยู่แล้ว
    maintenance_periods = [
        (pd.Timestamp('2021-04-10'), pd.Timestamp('2021-04-15')),
        (pd.Timestamp('2022-09-05'), pd.Timestamp('2022-09-10'))
    ]

    emergency_periods = []
    for date in ['2021-07-15', '2022-11-30']:
        start_date = pd.Timestamp(date)
        end_date = start_date + pd.Timedelta(hours=48)
        emergency_periods.append((start_date, end_date))

    # ทำเครื่องหมายเหตุการณ์พิเศษที่มีอยู่แล้ว
    for start, end in maintenance_periods + emergency_periods:
        mask = (df_anomaly.index >= start) & (df_anomaly.index <= end)
        df_anomaly.loc[mask, 'anomaly'] = 1

        # ระบุประเภทของความผิดปกติ
        if (start, end) in maintenance_periods:
            df_anomaly.loc[mask, 'anomaly_type'] = 'maintenance'
        else:
            df_anomaly.loc[mask, 'anomaly_type'] = 'emergency'

    # 1. เพิ่ม Point Anomalies
    # เลือกวันที่ไม่ใช่วันที่มีเหตุการณ์พิเศษอยู่แล้ว
    normal_indices = df_anomaly[df_anomaly['anomaly'] == 0].index

    if len(normal_indices) > 0:
        # สุ่มเลือกวันที่จะเพิ่ม point anomalies
        point_anomaly_indices = np.random.choice(normal_indices, n_point_anomalies, replace=False)

        for idx in point_anomaly_indices:
            # สุ่มว่าจะเพิ่มหรือลดค่า
            if np.random.rand() > 0.5:
                # เพิ่มค่า (spike)
                df_anomaly.loc[idx, 'power_production'] *= (1.5 + np.random.rand())  # เพิ่ม 150-250%
            else:
                # ลดค่า (dip)
                df_anomaly.loc[idx, 'power_production'] *= (0.3 + 0.2 * np.random.rand())  # ลดเหลือ 30-50%

            df_anomaly.loc[idx, 'anomaly'] = 1
            df_anomaly.loc[idx, 'anomaly_type'] = 'point'

    # 2. เพิ่ม Collective Anomalies
    # หาช่วงเวลาที่ไม่มีความผิดปกติติดกันอย่างน้อย 24 ชั่วโมง
    for _ in range(n_collective_anomalies):
        normal_series = df_anomaly[df_anomaly['anomaly'] == 0]
        if len(normal_series) < 24:
            break

        # หาจุดเริ่มต้นที่มีข้อมูลปกติติดกันอย่างน้อย 24 ชั่วโมง
        consecutive_normal = pd.Series(normal_series.index).diff() == pd.Timedelta(hours=1)
        start_points = []

        count = 0
        for i, is_consecutive in enumerate(consecutive_normal):
            if is_consecutive:
                count += 1
                if count >= 24:
                    start_points.append(i - 23)  # จุดเริ่มต้นของชุดข้อมูลปกติ 24 ชม.
            else:
                count = 0

        if len(start_points) == 0:
            continue

        # สุ่มเลือกจุดเริ่มต้น
        start_idx = np.random.choice(start_points, 1)[0]
        anomaly_length = np.random.randint(12, 25)  # ความยาว 12-24 ชั่วโมง

        # กำหนดรูปแบบของ collective anomaly
        anomaly_pattern = np.random.choice(['gradual_increase', 'oscillation', 'flat'])

        for i in range(anomaly_length):
            current_idx = normal_series.index[start_idx + i]

            if anomaly_pattern == 'gradual_increase':
                # เพิ่มขึ้นอย่างต่อเนื่อง
                df_anomaly.loc[current_idx, 'power_production'] *= (1 + 0.05 * i)
            elif anomaly_pattern == 'oscillation':
                # แกว่งขึ้นลง
                df_anomaly.loc[current_idx, 'power_production'] *= (1 + 0.2 * np.sin(i))
            elif anomaly_pattern == 'flat':
                # ค่าคงที่ผิดปกติ (ไม่มีรูปแบบรายวัน)
                avg_production = df_anomaly.loc[current_idx, 'power_production']
                df_anomaly.loc[current_idx, 'power_production'] = avg_production

            df_anomaly.loc[current_idx, 'anomaly'] = 1
            df_anomaly.loc[current_idx, 'anomaly_type'] = f'collective_{anomaly_pattern}'

    # 3. เพิ่ม Contextual Anomalies
    # ความผิดปกติตามบริบท เช่น ค่าสูงในช่วงกลางคืน (ที่ปกติควรต่ำ)
    for _ in range(n_contextual_anomalies):
        # เลือกช่วงกลางคืน (22:00 - 04:00) ที่เป็นข้อมูลปกติ
        night_hours = df_anomaly[(df_anomaly['hour'] >= 22) | (df_anomaly['hour'] <= 4)]
        night_normal = night_hours[night_hours['anomaly'] == 0]

        if len(night_normal) == 0:
            continue

        # สุ่มเลือกวันที่
        context_idx = np.random.choice(night_normal.index, 1)[0]

        # เพิ่มค่าผิดปกติในช่วงกลางคืน (ที่ปกติควรต่ำ)
        avg_day_production = df_anomaly[(df_anomaly['hour'] >= 10) & (df_anomaly['hour'] <= 16)]['power_production'].mean()
        df_anomaly.loc[context_idx, 'power_production'] = avg_day_production * (0.9 + 0.2 * np.random.rand())

        df_anomaly.loc[context_idx, 'anomaly'] = 1
        df_anomaly.loc[context_idx, 'anomaly_type'] = 'contextual_night_high'

    return df_anomaly

# ============================================================================
# PART 3: Data Preprocessing Functions
# ============================================================================

def create_sequences(data, seq_length):
    """
    แปลงข้อมูล time series เป็นชุด sequences สำหรับการเรียนรู้ของ LSTM
    """
    xs = []
    ys = []

    for i in range(len(data) - seq_length):
        x = data[i:(i + seq_length)]
        xs.append(x)

        # label คือค่า anomaly ของจุดสุดท้ายในลำดับ
        y = data.iloc[i + seq_length - 1]['anomaly']
        ys.append(y)

    return np.array(xs), np.array(ys)

def prepare_data_for_lstm(df, features, seq_length=24, test_size=0.2):
    """
    เตรียมข้อมูลสำหรับ LSTM Autoencoder
    """
    # Normalize ข้อมูล
    scaler = MinMaxScaler()
    df_scaled = df.copy()
    df_scaled[features] = scaler.fit_transform(df[features])

    # แยกข้อมูลปกติสำหรับฝึกโมเดล
    train_size = int(len(df_scaled) * (1 - test_size))
    df_train = df_scaled.iloc[:train_size].copy()
    df_test = df_scaled.iloc[train_size:].copy()

    # แยกเฉพาะข้อมูลปกติสำหรับการฝึกโมเดล
    df_train_normal = df_train[df_train['anomaly'] == 0].copy()

    # สร้าง sequences
    X_train, _ = create_sequences(df_train_normal[features + ['anomaly']], seq_length)
    X_test, y_test = create_sequences(df_test[features + ['anomaly']], seq_length)

    # ปรับรูปร่างข้อมูล: (samples, time steps, features)
    n_features = len(features)
    X_train = X_train[:, :, :n_features]  # ตัดคอลัมน์ anomaly ออก
    X_test = X_test[:, :, :n_features]    # ตัดคอลัมน์ anomaly ออก

    return X_train, X_test, y_test, scaler, df_train, df_test

# ============================================================================
# PART 4: Model Building and Evaluation Functions
# ============================================================================

def build_lstm_autoencoder(seq_length, n_features, encoding_dim=32):
    """
    สร้างโมเดล LSTM Autoencoder
    """
    # Define input
    inputs = Input(shape=(seq_length, n_features))

    # Encoder
    encoded = LSTM(64, activation='relu', return_sequences=True)(inputs)
    encoded = LSTM(encoding_dim, activation='relu')(encoded)

    # Bottleneck representation (latent space)
    # Decoder
    decoded = RepeatVector(seq_length)(encoded)
    decoded = LSTM(encoding_dim, activation='relu', return_sequences=True)(decoded)
    decoded = LSTM(64, activation='relu', return_sequences=True)(decoded)

    # Output
    outputs = TimeDistributed(Dense(n_features))(decoded)

    # Create model
    autoencoder = Model(inputs, outputs)
    autoencoder.compile(optimizer='adam', loss='mse')

    return autoencoder

def detect_anomalies(model, X, threshold=None, contamination=0.01):
    """
    ตรวจจับความผิดปกติโดยใช้ LSTM Autoencoder ที่ฝึกแล้ว
    """
    # ทำนายข้อมูล
    X_pred = model.predict(X)

    # คำนวณ Reconstruction Error (Mean Squared Error ระหว่างข้อมูลจริงและข้อมูลที่ทำนาย)
    mse = np.mean(np.square(X - X_pred), axis=(1, 2))

    # ถ้าไม่ได้กำหนด threshold ให้ใช้เปอร์เซ็นไทล์ตามค่า contamination
    if threshold is None:
        threshold = np.percentile(mse, 100 * (1 - contamination))

    # ระบุความผิดปกติ (1 = ผิดปกติ, 0 = ปกติ)
    anomalies = (mse > threshold).astype(int)

    return anomalies, mse, threshold

def evaluate_anomaly_detection(y_true, y_pred):
    """
    ประเมินประสิทธิภาพของการตรวจจับความผิดปกติ
    """
    precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, average='binary')

    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")

    # สร้าง Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show()

    return precision, recall, f1

# ============================================================================
# PART 5: Main Execution
# ============================================================================

# สร้างข้อมูล
print("Generating power production data...")
power_data = create_power_production_data()
print(f"Generated {len(power_data)} data points from {power_data.index.min()} to {power_data.index.max()}")

# เพิ่มความผิดปกติ
print("\nAdding anomalies to the data...")
power_data_with_anomalies = add_anomalies(power_data)

# สรุปข้อมูลความผิดปกติที่เพิ่มเข้าไป
anomaly_counts = power_data_with_anomalies['anomaly_type'].value_counts()
print("\nAnomaly Distribution:")
print(anomaly_counts)

# แสดงกราฟข้อมูล
plt.figure(figsize=(15, 8))
plt.plot(power_data_with_anomalies.index, power_data_with_anomalies['power_production'], label='Power Production')

# แสดงจุดที่เป็นความผิดปกติแยกตามประเภท
anomaly_types = power_data_with_anomalies[power_data_with_anomalies['anomaly'] == 1]['anomaly_type'].unique()
for anomaly_type in anomaly_types:
    subset = power_data_with_anomalies[power_data_with_anomalies['anomaly_type'] == anomaly_type]
    plt.scatter(subset.index, subset['power_production'], label=anomaly_type, alpha=0.8)

plt.title('Power Production with Anomalies')
plt.xlabel('Date')
plt.ylabel('Power Production (MW)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# ============================================================================
# PART 6: LSTM Autoencoder Model Training and Evaluation
# ============================================================================

# ฟีเจอร์ที่จะใช้
features = ['power_production']

# เตรียมข้อมูล
X_train, X_test, y_test, scaler, df_train, df_test = prepare_data_for_lstm(
    power_data_with_anomalies,
    features,
    seq_length=24,  # 24 ชั่วโมงย้อนหลัง
    test_size=0.3   # ใช้ 30% ของข้อมูลเป็นชุดทดสอบ
)

print(f"\nTraining data shape: {X_train.shape}")
print(f"Testing data shape: {X_test.shape}")
print(f"Test labels shape: {y_test.shape}")

# สร้างและฝึกโมเดล
model = build_lstm_autoencoder(
    seq_length=24,
    n_features=len(features),
    encoding_dim=16  # มิติของ latent space
)

print("\nModel Summary:")
model.summary()

# Early Stopping เพื่อป้องกัน overfitting
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# ฝึกโมเดล
print("\nTraining LSTM Autoencoder...")
history = model.fit(
    X_train, X_train,  # Autoencoder เรียนรู้การสร้างข้อมูลเดิม
    epochs=50,
    batch_size=32,
    validation_split=0.1,
    callbacks=[early_stopping],
    verbose=1
)

# แสดงกราฟการฝึก
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss During Training')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()

# ตรวจจับความผิดปกติ
print("\nDetecting anomalies...")
y_pred, mse, threshold = detect_anomalies(model, X_test)

# ประเมินผล
print("\nModel Evaluation:")
precision, recall, f1 = evaluate_anomaly_detection(y_test, y_pred)

# แสดงกราฟ Reconstruction Error
plt.figure(figsize=(15, 6))
plt.plot(mse, label='Reconstruction Error')
plt.axhline(y=threshold, color='r', linestyle='--', label=f'Threshold: {threshold:.6f}')
plt.title('Reconstruction Error for Test Data')
plt.xlabel('Data Point')
plt.ylabel('Reconstruction Error (MSE)')
plt.legend()
plt.grid(True)
plt.show()

# ============================================================================
# PART 7: Visualize Results on Original Time Series
# ============================================================================

# สร้างดัชนีเพื่อจับคู่ผลการทำนายกับข้อมูลดิบ
test_indices = df_test.index[23:len(df_test)-1]  # ปรับเนื่องจาก sequence_length=24 และการสร้าง sequences

# สร้าง DataFrame สำหรับแสดงผล
results_df = pd.DataFrame({
    'timestamp': test_indices,
    'power_production': df_test.loc[test_indices, 'power_production'],
    'true_anomaly': y_test,
    'predicted_anomaly': y_pred,
    'reconstruction_error': mse,
    'threshold': threshold
})

# แสดงกราฟผลลัพธ์
plt.figure(figsize=(15, 10))

# กราฟกำลังการผลิตไฟฟ้าและความผิดปกติ
plt.subplot(2, 1, 1)
plt.plot(results_df['timestamp'], scaler.inverse_transform(results_df[['power_production']]), label='Power Production')

# แสดงความผิดปกติจริง
true_anomalies = results_df[results_df['true_anomaly'] == 1]
plt.scatter(true_anomalies['timestamp'],
           scaler.inverse_transform(true_anomalies[['power_production']]),
           color='red', label='True Anomaly', alpha=0.7, marker='o')

# แสดงความผิดปกติที่ทำนาย
predicted_anomalies = results_df[results_df['predicted_anomaly'] == 1]
plt.scatter(predicted_anomalies['timestamp'],
           scaler.inverse_transform(predicted_anomalies[['power_production']]),
           color='green', label='Predicted Anomaly', alpha=0.7, marker='x', s=100)

plt.title('Power Production with Anomaly Detection Results')
plt.xlabel('Date')
plt.ylabel('Power Production (MW)')
plt.legend()
plt.grid(True)

# กราฟแสดง Reconstruction Error
plt.subplot(2, 1, 2)
plt.plot(results_df['timestamp'], results_df['reconstruction_error'], label='Reconstruction Error')
plt.axhline(y=threshold, color='r', linestyle='--', label=f'Threshold: {threshold:.6f}')
plt.title('Reconstruction Error Over Time')
plt.xlabel('Date')
plt.ylabel('Reconstruction Error (MSE)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

# ============================================================================
# PART 8: Comparison with Different Feature Sets
# ============================================================================

print("\n\nExperimenting with a multivariate approach...")

# สร้างฟีเจอร์เพิ่มเติม
power_data_with_anomalies['hour_sin'] = np.sin(2 * np.pi * power_data_with_anomalies['hour'] / 24)
power_data_with_anomalies['hour_cos'] = np.cos(2 * np.pi * power_data_with_anomalies['hour'] / 24)
power_data_with_anomalies['day_sin'] = np.sin(2 * np.pi * power_data_with_anomalies['day_of_week'] / 7)
power_data_with_anomalies['day_cos'] = np.cos(2 * np.pi * power_data_with_anomalies['day_of_week'] / 7)

# ฟีเจอร์ที่รวมข้อมูลเวลา
multivariate_features = ['power_production', 'hour_sin', 'hour_cos', 'day_sin', 'day_cos']

# เตรียมข้อมูลเพิ่มเติม
X_train_multi, X_test_multi, y_test_multi, _, _, _ = prepare_data_for_lstm(
    power_data_with_anomalies,
    multivariate_features,
    seq_length=24,
    test_size=0.3
)

print(f"\nMultivariate training data shape: {X_train_multi.shape}")
print(f"Multivariate testing data shape: {X_test_multi.shape}")

# สร้างและฝึกโมเดล multivariate
model_multi = build_lstm_autoencoder(
    seq_length=24,
    n_features=len(multivariate_features),
    encoding_dim=16
)

# ฝึกโมเดล
print("\nTraining Multivariate LSTM Autoencoder...")
history_multi = model_multi.fit(
    X_train_multi, X_train_multi,
    epochs=50,
    batch_size=32,
    validation_split=0.1,
    callbacks=[early_stopping],
    verbose=1
)

# ตรวจจับความผิดปกติ
print("\nDetecting anomalies with multivariate model...")
y_pred_multi, mse_multi, threshold_multi = detect_anomalies(model_multi, X_test_multi)

# ประเมินผล
print("\nMultivariate Model Evaluation:")
precision_multi, recall_multi, f1_multi = evaluate_anomaly_detection(y_test_multi, y_pred_multi)

# เปรียบเทียบผลลัพธ์
print("\nComparison:")
print(f"Univariate F1 Score: {f1:.4f}")
print(f"Multivariate F1 Score: {f1_multi:.4f}")
print(f"Improvement: {(f1_multi-f1)/f1*100:.2f}%")

print("\n\nExperiment completed! You can modify the hyperparameters, try different architectures, or explore other feature combinations.")
```
